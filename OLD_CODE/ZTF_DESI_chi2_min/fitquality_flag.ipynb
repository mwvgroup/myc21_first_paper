```python
import pandas as pd

# Load your CSV file
df = pd.read_csv("/Users/pittsburghgraduatestudent/repos/myc21_ztf_mu/ZTF_snia_data.csv")

df.replace({"sub_type": {None: "nan"}}, inplace=True)

# Apply both filters
# THis classifies as suitable for SN cosmology As stated in the paper right before ch. 4

filtered_df = df[
    (df["fitquality_flag"] == 1) & 
    (df["lccoverage_flag"] == 1) &
    (df["sub_type"] == "norm") |
    (df["sub_type"] == "nan")
]

# Print summary
print(f"Original rows: {len(df)}")
print(f"Filtered rows: {len(filtered_df)}")

print(filtered_df.groupby("sub_type").size())
print(filtered_df["sub_type"].unique())

filtered_df.to_csv("ZTF_data_good_4_cosmology.csv", index=False)



```
